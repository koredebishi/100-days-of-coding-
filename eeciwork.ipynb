```python
import numpy as np
import pandas as pd
import openpyxl



workbook = openpyxl.load_workbook('/content/eecs_work.xlsx')

input_sheet = 'Working_document'
eecs_Input = pd.read_excel('/content/eecs_work.xlsx', sheet_name=input_sheet)

eecs_output = pd.DataFrame()
for index, row in eecs_Input.iterrows():
  input_columnA = row['Q10_Confidence']

  if input_columnA in ['Quite confident', 'Extremely confident']:
      result = 1
  elif input_columnA in ['Not confident at all', 'Somewhat confident', 'Slightly confident']:
      result = 0
  else:
      result = None

  eecs_output = eecs_output.append({'Q10 C-Level': result}, ignore_index=True)

output_sheet= 'Version_encode'
with pd.ExcelWriter('/content/eecs_work.xlsx', engine='openpyxl') as writer:
    writer.book = workbook
    eecs_output.to_excel(writer, sheet_name=output_sheet, index=False)

workbook.save('/content/eecs_work.xlsx')

```
